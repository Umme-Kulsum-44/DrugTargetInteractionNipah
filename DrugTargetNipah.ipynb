```python
from sklearn.metrics import confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np

def classification(X_train, X_test, y_train, y_test):
    # Encode the labels to numeric values
    label_mapping = {'Active': 1, 'Inactive': 0}
    y_train = y_train.map(label_mapping)
    y_test = y_test.map(label_mapping)

    # Define classification algorithms
    Classifiers = {
        'Support Vector Machine': SVC(probability=True),
        'Logistic Regression': LogisticRegression(),
        'K-Nearest Neighbors': KNeighborsClassifier(),
        'Naive Bayes': GaussianNB(),
        'Decision Tree': DecisionTreeClassifier(),
        'Random Forest': RandomForestClassifier(),
        'Gradient Boosting': GradientBoostingClassifier()
    }

    for name, clf in Classifiers.items():
        print(name)
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        y_proba = clf.predict_proba(X_test)[:, 1] if hasattr(clf, "predict_proba") else clf.decision_function(X_test)

        # Calculate accuracy
        accuracy = accuracy_score(y_test, y_pred)
        print(f"{name} - Accuracy: {accuracy:.4f}")

        # Calculate specificity and sensitivity
        tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
        specificity = tn / (tn + fp)
        sensitivity = tp / (tp + fn)
        print(f"{name} - Specificity: {specificity:.4f}")
        print(f"{name} - Sensitivity: {sensitivity:.4f}")

        # Print classification report
        print(classification_report(y_test, y_pred))
        print()

        # Display confusion matrix
        cm = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, cmap='Blues', fmt='g')
        plt.title(f"Confusion Matrix for {name}")
        plt.xlabel("Predicted")
        plt.ylabel("Actual")
        plt.show()

        # Compute ROC curve and area under the curve
        fpr, tpr, thresholds = roc_curve(y_test, y_proba)
        roc_auc = auc(fpr, tpr)

        # Plot ROC curve
        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, color='darkorange', lw=2, label=f"ROC curve (area = {roc_auc:.2f})")
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve for {name}')
        plt.legend(loc="lower right")
        plt.show()
```

```python
# Map 'Active' and 'Inactive' to integers
label_mapping = {'Inactive': 0, 'Active': 1}
y = y.map(label_mapping)

# Convert to NumPy array (if y is still a Pandas Series)
y = y.values

# Convert y to categorical for binary classification
y = to_categorical(y, num_classes=2)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create CNN model
model = Sequential([
    Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], 1)),
    MaxPooling1D(pool_size=2),
    Dropout(0.3),
    Conv1D(filters=128, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    Dropout(0.3),
    Flatten(),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(2, activation='softmax')  # Output layer for binary classification
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))

# Evaluate the model
y_pred = np.argmax(model.predict(X_test), axis=1)
y_test_labels = np.argmax(y_test, axis=1)

# Metrics
accuracy = accuracy_score(y_test_labels, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_test_labels, y_pred))
```

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, LSTM, GRU, Flatten, Dense, Dropout, concatenate
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import numpy as np

# Ensure X_balanced is in the correct shape (samples, timesteps, features)
X_balanced = X_balanced.reshape(X_balanced.shape[0], X_balanced.shape[1], 1)

# Convert y_balanced to categorical (binary classification: Active=1, Inactive=0)
y_balanced = to_categorical(y_balanced, num_classes=2)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_balanced, y_balanced, test_size=0.3, random_state=42)

# CNN Model
cnn_input = Input(shape=(X_train.shape[1], X_train.shape[2]))
cnn = Conv1D(filters=64, kernel_size=3, activation='relu')(cnn_input)
cnn = MaxPooling1D(pool_size=2)(cnn)
cnn = Dropout(0.3)(cnn)
cnn = Flatten()(cnn)

# RNN Model
rnn_input = Input(shape=(X_train.shape[1], X_train.shape[2]))
rnn = LSTM(64, return_sequences=True)(rnn_input)
rnn = Dropout(0.3)(rnn)
rnn = GRU(64, return_sequences=False)(rnn)
rnn = Dropout(0.3)(rnn)

# Concatenate CNN and RNN Outputs
merged = concatenate([cnn, rnn])
merged = Dense(64, activation='relu')(merged)
merged = Dropout(0.3)(merged)
output = Dense(2, activation='softmax')(merged)  # Output layer for binary classification

# Define the model
model = Model(inputs=[cnn_input, rnn_input], outputs=output)

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit([X_train, X_train], y_train, epochs=10, batch_size=32, validation_data=([X_test, X_test], y_test))

# Evaluate the model
y_pred = np.argmax(model.predict([X_test, X_test]), axis=1)
y_test_labels = np.argmax(y_test, axis=1)

# Metrics
accuracy = accuracy_score(y_test_labels, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_test_labels, y_pred))

```
